# Reading from file function

```python
def read_automata(path):
    file = open(path, 'r')
    line = file.readline()
    states = line.split()
    line = file.readline()
    abc = line.split()
    line = file.readline()
    start_sates = line.split()
    line = file.readline()
    end_states = line.split()
    transitions = []
    for line in file:
        transition = line.split()
        if len(transition) != 3:
            raise RuntimeError('file read error')
        transitions.append(tuple(transition))
    return Automata(states, abc, start_sates, end_states, transitions)
```

# Reading from file

```python
at = read_automata('input.txt')
```
